```python
import os
import pandas as pd
import datetime as dt
import json
```

```python
path = '../data/csvs/'
filename = '20240904_08_28_96_max_jac_dmn_as_cntry_pyasn_org_edited.csv'
df = pd.read_csv(path + filename)
df.head()
```

```python
# Extract the date part (first 8 characters before the first '_')
date_str = filename.split('_')[0]  # e.g., '20240904'

# Convert to datetime object (assuming YYYYMMDD format)
date_obj = dt.datetime.strptime(date_str, '%Y%m%d')

timestamp = int(date_obj.timestamp())

df['timestamp'] = timestamp

# Optional: Display a sample to verify
print(f"Extracted timestamp: {timestamp} (corresponds to {date_obj})")
df.head()
```

```python
all_data = []

csv_files = [f for f in os.listdir(path) if f.endswith('.csv')]

for filename in csv_files:
    # Extract the date part (first 8 characters before the first '_')
    date_str = filename.split('_')[0]  # e.g., '20240904'
    
    # Convert to datetime object (assuming YYYYMMDD format)
    date_obj = dt.datetime.strptime(date_str, '%Y%m%d')
    
    timestamp = int(date_obj.timestamp())
    
    df = pd.read_csv(os.path.join(path, filename))
    
    df['timestamp'] = timestamp
    
    all_data.extend(df.to_dict('records'))

with open('../data/jsons/combined_sibling_prefixes_data.json', 'w') as f:
    json.dump(all_data, f, indent=4)

print(f"Combined data from {len(csv_files)} files saved to combined_sibling_prefixes_data.json")
```
